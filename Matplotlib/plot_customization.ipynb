```python
import matplotlib.pyplot as plt
import numpy as np


x =np.array(["Salohiddin","Jacky","Adam","Steve"])
y = np.array([10,30,40,20])
y1 = np.array([30,10,0,20])
y2 = np.array([20,20,20,20])


plt.plot(x,y,marker=".",ms=25,
         markerfacecolor="yellow",
         markeredgecolor="yellow",
         linestyle = "solid",
         linewidth=4,
         color="red")
plt.plot(x,y1,marker=".",ms=25,
         markerfacecolor="yellow",
         markeredgecolor="yellow",
         linestyle = "solid",
         linewidth=4,
         color="green")
plt.plot(x,y2,marker=".",ms=25,
         markerfacecolor="yellow",
         markeredgecolor="yellow",
         linestyle = "solid",
         linewidth=4,
         color="orange")
plt.show()
```
