```python
import matplotlib.pyplot as plt
import numpy as np


x =np.array([2023,2024,2025,2026])
y1 = np.array([10,30,40,20])
y2 = np.array([30,10,0,20])
y3 = np.array([20,20,20,20])

plt.title("Class size",fontsize=20,
          family="Arial",
          fontweight="bold",
          color="blue")

plt.xlabel("Year",fontsize=15,
           family="Arial",
           fontweight="bold",
           color="red"
           )
plt.ylabel("Students",fontsize=15,
           family="Arial",
           fontweight="bold",
           color="black"
           )

plt.xticks(x)
plt.tick_params("both",
                colors="blue")


plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create BPSK diagram
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# BPSK
axes[0].scatter([1, -1], [0, 0], s=200)
axes[0].axhline(y=0, color='k', linestyle='--', alpha=0.3)
axes[0].axvline(x=0, color='k', linestyle='--', alpha=0.3)
axes[0].set_xlim(-1.5, 1.5)
axes[0].set_ylim(-1.5, 1.5)
axes[0].set_title('BPSK Constellation')
axes[0].set_xlabel('I')
axes[0].set_ylabel('Q')
axes[0].grid(True)

# QPSK
points = [(1/np.sqrt(2), 1/np.sqrt(2)), (-1/np.sqrt(2), 1/np.sqrt(2)),
          (-1/np.sqrt(2), -1/np.sqrt(2)), (1/np.sqrt(2), -1/np.sqrt(2))]
x, y = zip(*points)
axes[1].scatter(x, y, s=200)
axes[1].axhline(y=0, color='k', linestyle='--', alpha=0.3)
axes[1].axvline(x=0, color='k', linestyle='--', alpha=0.3)
axes[1].set_xlim(-1.5, 1.5)
axes[1].set_ylim(-1.5, 1.5)
axes[1].set_title('QPSK Constellation')
axes[1].set_xlabel('I')
axes[1].set_ylabel('Q')
axes[1].grid(True)

# Comparison
axes[2].bar(['ASK', 'BPSK', 'QPSK'], [1, 2, 4])
axes[2].set_title('Bits per Symbol Comparison')
axes[2].set_ylabel('Bits/Symbol')

plt.tight_layout()
plt.savefig('constellation_diagrams.png', dpi=300)
plt.show()
```
