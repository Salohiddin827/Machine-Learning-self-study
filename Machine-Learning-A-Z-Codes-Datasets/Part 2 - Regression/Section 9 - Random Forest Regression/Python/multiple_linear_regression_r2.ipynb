# Multiple Linear Regression with R² and Adjusted R²
This notebook demonstrates how to train a multiple linear regression model and compute R² and Adjusted R².

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

## Create Example Dataset

```python
data = {
    'Size':[120,150,90,200,175,160,140,210],
    'Rooms':[3,4,2,5,4,4,3,5],
    'Distance':[5,3,8,2,4,3,6,1],
    'Price':[300,450,200,600,520,480,350,650]
}

dataset = pd.DataFrame(data)
dataset
```

## Split Features and Target

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

## Train/Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)
```

## Train Multiple Linear Regression Model

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

## Make Predictions

```python
y_pred = model.predict(X_test)
y_pred
```

## Compute R²

```python
r2 = r2_score(y_test, y_pred)
print('R2:', r2)
```

## Compute Adjusted R²

```python
n = X_test.shape[0]
p = X_test.shape[1]

adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
print('Adjusted R2:', adjusted_r2)
```

## Model Parameters

```python
print('Intercept:', model.intercept_)
print('Coefficients:', model.coef_)
```
