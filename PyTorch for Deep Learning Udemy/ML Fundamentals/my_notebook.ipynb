```python
#reshaping, stacking squeezing and unsqueezing tensors
```

Reshaping -- reshapes an input tensor to a defined shape


View -- Return a view of an input tensor of certain shape but keep the same memory as the original tensor

Stacking -- combine multiple tensors on top of each other (vstack) or side by side (hstack)

Squeeze -- removes all 'l' dimension to a target tensor

Unsequeeze -- add a 1 dimensiopn to a target tensor

Permute -- return aa view pf the input with dimensions permuted(Swapped) in 
 certain way

```python
#Let's create a tensor 

import torch
x = torch.arange(1.,10.)
x, x.shape
```

```python
#Add a extra dimension 
x_reshaped = x.reshape(9,1)
x_reshaped, x_reshaped.shape
```

```python
#Change the view 
z = x.view(1,9)
z, z.shape
```

```python
#changeing z changes x (because a aview of a  tensor shares the same memory as the original input)
z[:,0] = 5
z,x
```

```python
#Stack tensors on top of each other
x_stacked  = torch.stack([x,x,x,x],dim = 1)
x_stacked
```

```python
#torch.squeeze () -- removes all single dimesnions from a target tensor
print(f"Previous tensor {x_reshaped}")
print(f"Previous shape {x_reshaped.shape}")


#Remove an extra dimensions form a x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"New {x_squeezed}")
print(f"New shape {x_squeezed.shape}")
```

```python
#torch.unsqueeze adds a single dimension to target tensor at a specific tim (dimension)
print(f"Previous target {x_squeezed}")
print(f"Previous shape {x_squeezed.shape}")

x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"New target {x_squeezed}")
print(f"New shape {x_squeezed.shape}")
```

```python
#torch.permute -- rearranges the dimensions of a atarget tensor in a specified order
x_original = torch.rand(size =(224,224,3)) #height width chanel colour

#Permute the originial tensor to rearrange the axis (or dim) order

x_permuted = x_original.permute(2,0,1)
print(f"Previous shape  {x_original}")
print(f"New shape {x_permuted.shape}")

x_original[0,0,0] = 7 
x_original

```
