```python
import pandas as pd
```

```python
#Series
values = [10,20,30,40,50]
s= pd.Series(values,index=['a','b','c','d','e'])
d = s.loc['a']
print(s)
print(d)
```

```python
# Data Frames
df = pd.DataFrame({
    'name': ['Mike','Bob','Alice'],
    'age':[30,40,50],
    'job':['Programmer','Clerk','Designer']

})
df = df.set_index('name')
mike_loc = df.loc['Mike']
print(mike_loc)
```

```python
df1 = pd.DataFrame({
    'a': [1, 2, 3]
}, index = [1,0,2])

df2 = pd.DataFrame({
    'a': [10, 20, 30]
},index = [1,2,0])
print(df1*df2)
```
